# Goal: understand pairwise distance metrics and how they are calculated
Using scipy pdist method and sklearn pairwise_distances method

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics import pairwise_distances

from scipy.linalg import norm
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

### Load Iris Data Set into Python

Each sample in $X$ is a row. 
Each column in $X$ is one of four features in centimeters:
* Sepal Length
* Sepal Width
* Petal Length
* Petal Width



```python
iris =  load_iris()  # load iris data set packagea
X = iris.data   #  get data only
X.shape  
```

150 samples with 4 features each

Let's use a very simple matrix to understand pairwise spatial metrics.

A matrix named $D$ is formed from 3 row rectors $a, b, c$

Each row $a, b, c$ represents a sample.

Each sample, for example $a$, has 4 variables or characteristics, just like our Iris Data set.

${a, b, c} \in \mathbb{R}^4$

```python
a = np.array([1, 0, 3, 1])  # sample a
b = np.array([1, -1, 5, 3])  # sample b
c = np.array([3, 4, 4, 3])   # sample c
d = np.array([a, b, c])

print('Matrix d: \n',d, '\n')
```

We can use different metrics for measuring the distance between vectors. Here are a 2 common ones:

Manhattan is the  $\rightarrow$  $L_1$ norm

Euclidean is the $\rightarrow$  $L_2$ norm


```python
M_dist_sk_man = pairwise_distances(d,metric='manhattan')  
M_dist_sk_ecl = pairwise_distances(d,metric='euclidean')  # euclidean

print("M_manhattan =\n", M_dist_sk_man, '\n\n', "M_euclidean =\n", M_dist_sk_ecl)
```

### All of a sudden we get a $3\rm{x}3$ Distance Matrix
What do the elements of the distance matrix tell us? $n$ refers to the $n^{th}$ norm:
Where $n=2$ refers to the  $L_2$ norm.

![M_norm](M_norm.png)


The diagonal of $M_{manhattan}$ and $M_{euclidean}$ are 0 due to pair-wise distance of vectors with itself.

For example: $M_{11}$ is the pair wise distance vector of $a$ and itself: $||a-a||_n$,  $M_{22}$ is the pair wise distance vector of $b$ and itself: $||b-b||_n$  

Remember, regardless of the norm, the norm of any vector and itself has to be 0

```python
print('M_11 =' ,norm(a-a,1))  # where 1 is the l1 Norm  
print('M_22 =' ,norm(b-b,1))
print('M_33 =' ,norm(c-c,1))
```

### What about the off diagonal elements?
Let's start with the $L_1$ norm
![M_manhattan](M_L_1.png)

```python
print('M_12(L_1 Norm) =', norm(a-b, 1))
print('M_13(L_1 Norm) =', norm(a-c, 1))
print('M_23(L_1 Norm) =', norm(b-c, 1))
```

Due to symmetry $||a-b||_n$ = $||b-a||_n$, this holds for all pair-wise elements. 

### This is why the matrix returned is symmetric!

If you're more familliar with the correlation matrix, one can think of the off diagonials as the "correlated" distance between the rows of the matrix $D$ symmetric. Both matricies are also symmetric.

![M_norm_1](M_L_2.png)

```python
print('M_12(L_2 Norm) =', norm(a-b, 2))
print('M_13(L_2 Norm) =', norm(a-c, 2))
print('M_23(L_2 Norm) =', norm(b-c, 2))
```

### Now! Let's return to the Iris Dataset

Since pairwise_distances method from sklearn returns a symmetric matrix, all the information is either in the upper or lower diagonal of the matrix.  

```python
X_dist_sk_manh = pairwise_distances(X,metric='manhattan')
X_dist_sk_eucl = pairwise_distances(X,metric='euclidean')
print("SKlearn Return Matrix Size: ",X_dist_sk_eucl.shape)

upper_tri_X_eucl = X_dist_sk_eucl[np.triu_indices(X_dist_sk_eucl.shape[0], k=1)]  # k=1 doesn't include diagonal
print("SKlearn Flattened array Size: ",upper_tri_X_eucl.shape)
```

Similarly, we can also use scipy's pdist method which automatically gets the lower or upper triangular values and flattens 

```python
X_dist = pdist(X)  #  euclidean
print('Scipys returned array:' ,X_dist.shape)
```

### Let's make a heatmap from the Distance Matrix

```python
mask = np.zeros_like(X_dist_sk_eucl)
mask[np.triu_indices_from(mask)] = True
fig, axe = plt.subplots(figsize=(15,15))
axe = sns.heatmap(X_dist_sk_eucl, mask=mask, square=True, cbar_kws={'label': 'My Colorbar'})
cbar = axe.collections[0].colorbar
plt.ylabel('Iris Dataset Samples', labelpad=20, fontsize=15)
plt.xlabel('Iris Dataset Samples', labelpad=20, fontsize=15)
cbar.set_label(r'Pairwise Distance: $L_{2}$ Norm', labelpad=20, fontsize=20)
```
